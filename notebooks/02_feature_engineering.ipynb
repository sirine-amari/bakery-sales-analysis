```python
import pandas as pd
import numpy as np
```

```python
# 1) Charger
df = pd.read_csv("../data/raw/daily_sales_french_bakery.csv")

# 2) Date
df["ds"] = pd.to_datetime(df["ds"])
df = df.sort_values("ds").set_index("ds")

# 3) Série globale journalière
daily_sales = df.groupby(df.index)["y"].sum().asfreq("D")
daily_sales = daily_sales.fillna(0)  # si jours manquants → 0 (ou méthode différente plus tard)

daily_sales.head()
```

```python
data = daily_sales.to_frame(name="y")
data.head()
```

```python
data["day_of_week"] = data.index.dayofweek          # 0=lundi ... 6=dimanche
data["month"] = data.index.month
data["day_of_month"] = data.index.day
data["week_of_year"] = data.index.isocalendar().week.astype(int)
data["is_weekend"] = (data["day_of_week"] >= 5).astype(int)

data.head()
```

```python
data["lag_1"] = data["y"].shift(1)
data["lag_7"] = data["y"].shift(7)
data["lag_14"] = data["y"].shift(14)
```

```python
data["roll_mean_7"] = data["y"].rolling(7).mean()
data["roll_std_7"] = data["y"].rolling(7).std()
data["roll_mean_30"] = data["y"].rolling(30).mean()
```

```python
data = data.dropna()
data.head(), data.shape
```

```python
data.to_csv("../data/processed/features_daily_sales.csv")
print("Fichier sauvegardé : data/processed/features_daily_sales.csv")
```
